```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
import pandas as pd

# path to your file in Fabric
file_path = "abfss://136a6b91-7b1c-4a1a-a8c0-9e81cf77e07c@onelake.dfs.fabric.microsoft.com/3ca45b72-dee5-4e97-b74a-d90763b17d54/Files/Sample - Superstore.csv"

df = pd.read_csv(file_path, encoding="latin1")

df.head()
```

```python
# Remove nulls
df = df.dropna()

df.head()

# Convert date
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
# Create new column
df['Year'] = df['Order Date'].dt.year

# Profit margin
df['Profit Margin'] = df['Profit'] / df['Sales']
df['Profit Margin'] = df['Profit Margin'].round(2)

# Sales category - segmentation
df['Sales Category'] = pd.cut(
df['Sales'],
bins=[0, 100, 500, 1000, 10000],
labels=['Low', 'Medium', 'High', 'Very High']
)

df.head()
```

```python
# Clean column names (VERY IMPORTANT)
df.columns = df.columns.str.replace(" ", "_")

df.head()
```

CREATE DIM_PRODUCT WITH SURROGATE KEY (duplicated prod id OFF-PA-10000477 with different prod names)

```python
# Create dimension table
dim_product = df[
    ['Product_ID', 'Product_Name', 'Category']
].drop_duplicates().reset_index(drop=True)

# Create surrogate key
dim_product['Product_Key'] = dim_product.index + 1

# Reorder columns
dim_product = dim_product[
    ['Product_Key', 'Product_ID', 'Product_Name', 'Category']
]

dim_product.head()
```

Merge tables -ADD PRODUCT_KEY TO FACT TABLE

```python
fact_sales = df.merge(
    dim_product,
    on=['Product_ID'],
    how='left'
)

fact_sales.head()
```

```python
# Create dim_product
dim_product = df[
    ['Product_ID', 'Product_Name', 'Category']
].drop_duplicates().reset_index(drop=True)

# Add surrogate key
dim_product['Product_Key'] = dim_product.index + 1

# Merge surrogate key into fact table
fact_sales = df.merge(
    dim_product[['Product_Key', 'Product_ID', 'Product_Name']],
    on=['Product_ID', 'Product_Name'],
    how='left'
)

fact_sales.head()
```
